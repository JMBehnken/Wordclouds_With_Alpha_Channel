# Generates Wordclouds with random letters in a rectangle

This notebook can be used to understand the basic functionalities of the python wordcloud library.

```python
#Import
import fuzz
import string
import random
import wordcloud as wc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def random_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None):
    #Set a random color
    h = int(360.0 * tone / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(70, 120)) / 255.0)
    return "hsl({}, {}%, {}%)".format(h, s, l)

tone = 10.0
```

```python
#Initiate the wordcloud
wordcloud = wc.WordCloud(width=2000,height=600, background_color='black', 
                      max_words=1628,relative_scaling=1,
                      color_func = random_color_func,
                      normalize_plurals=False)
```

```python
num_words = 100
min_letters = 3
max_letters = 10
min_size = 1
max_size = 10

#Get random words with random frequency
words = {''.join(random.choice(random.choice(string.ascii_uppercase)) for _ in range(random.randint(min_letters, max_letters))).capitalize():random.randint(min_size, max_size) for _ in range(num_words)}
```

```python
fig = plt.figure(1, figsize=(18,13))
ax1 = fig.add_subplot(1,1,1)

wordcloud.generate_from_frequencies(words)
ax1.imshow(wordcloud, interpolation="bilinear")
ax1.axis('off')
```

```python
#Save to file
wordcloud.to_file('build/Rectangle_Random_Letters.jpg')
wordcloud.to_file('build/Rectangle_Random_Letters.png')
```
